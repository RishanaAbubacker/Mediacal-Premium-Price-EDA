#**IV. Model Development and Feature Importances**

**A. Dataset Splitting**

Splitting the dataset into train set and test set in the ratio of 75:25 respectively

```python
data.columns
```

```python
data.dtypes
```

```python
# Features for Model development
X=data.drop(['PremiumPrice'], axis=1)
y=pd.DataFrame(data['PremiumPrice'])
```

```python
# Splitting the Datset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=42, test_size=0.25)
```

**B. Machine Learning Algorithms**

Fitting the training set into machine learing models to predict the nature of side effcets, according to various features

```python
# Importing evaluation metrics
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

**1. Random Forest Algorithm**

```python
# Fitting the training data and making the predictions on test data
from sklearn.ensemble import RandomForestRegressor
Random_forest=RandomForestRegressor()
Random_forest.fit(X_train,y_train)
y_pred = Random_forest.predict(X_test)
```

Observation:  

1. 'Age' is the most important factor in determining 'PremiumPrice'.
2. 'numberOfMajorSurgeries' and 'Weight' are slightly correlated with 'PremiumPrice'.
3. Rest of attributes have negligeble correlation with 'PremiumPrice'


```python
# Bar plot of feature importances for Random Forest Regressor
feature_imp1 = Random_forest.feature_importances_
sns.barplot(x=feature_imp1, y=X.columns)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title('Feature Importances')
plt.show()
```

**2. Extreme Gradient Boosting**

```python
# Fitting the training data and making the predictions on test data
from xgboost import XGBRFRegressor
xg=XGBRFRegressor()
xg.fit(X_train,y_train)
xg_predict=xg.predict(X_test)
```

```python
# Bar plot of feature importances for XG Boost Regressor
feature_imp2 = xg.feature_importances_
fig=sns.barplot(x=feature_imp2, y=X.columns)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title('Feature Importances')
plt.show()
```

Observation:  

1. 'Age' is the most important factor in determining 'PremiumPrice'.
2. 'numberOfMajorSurgeries' and 'AnyTransplants' are slightly correlated with 'PremiumPrice'.
3. Rest of attributes have negligeble correlation with 'PremiumPrice'


**Conclusion of analysis**
1. Age is the most important Factor followed by AnyTransplant, Weight and other.
2. KnownAllergies,Diabetes, BloodPressureProblems are the least important Factor in determining the Premium Price.
